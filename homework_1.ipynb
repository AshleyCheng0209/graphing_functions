# 1. $f(x) = \sin(x) + \cos(x) $
- ## 先畫出$f(x) = \sin(x)$跟$f(x) = \cos(x)$ 的圖形，再加總做比對第一張圖
- ## 最大值跟最小值為

## f(x) = sin(x) 跟 f(x) = cos(x) 的圖形
- ## 兩者最大值為1，最小值為-1
- ## f(x)=sin(x)交於原點(0,0)

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-3*np.pi,3*np.pi,200)
y = np.sin(x)
y1 = np.cos(x) 
f = lambda x: np.sin(x)+np.cos(x)
plt.plot(x,y,color='r')
plt.plot(x,y1,color='g')
plt.grid()
plt.legend(['sin(x)','cos(x)'])
plt.axis([-3*np.pi, 3*np.pi, -1.5, 1.5])
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.show()
```

## f(x)=sin(x)+cos(x)的圖
- ## 最大值為1.41，最小值為-1.41
- ## 與y軸交於(0,1)

```python
plt.plot(x,f(x),color='b',label='sin(x)+cos(x)')
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-3*np.pi, 3*np.pi, -1.5, 1.5])
plt.title('$f(x) = \sin(x) + \cos(x)$')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.show()
```

# 2. $f(x) = \frac{1-e^{-2x}}{1+e^{-2x}}$
- ## 漸近線為 y = 1 和 y = -1 且有一解在原點(0,0)

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-10,10,200)
y = (1-np.exp((-2)*x))/(1+np.exp((-2)*x))
plt.plot(x,y,color='b')
plt.grid(True)
plt.title(r'$f(x) = \frac{1-e^{-2x}}{1+e^{-2x}} $')
ax = plt.gca()
ax.spines['bottom'].set_position(('data',0))
plt.show()
```

# 3. $f(x) = \sqrt[3]{\frac{4-x^3}{1+x^2}}$
- ## 與y軸交於(0,1.59)

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = np.sign(((4-x**3)/(1+x**2)))*(np.abs(((4-x**3)/(1+x**2))))**(1/3)
plt.plot(x,y,color='b')
plt.grid(True)
plt.title(r'$f(x) = \sqrt[3]{\frac{4-x^3}{1+x^2}}$')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.show()
```

# 4. $f(x) = \frac{1}{x}$
- ## 當x趨向0，左極限和右極限分別趨向負無限及正無限，因此函數在x=0的極限不存在
- ## 漸進線為x跟y軸

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-10,10,101)
y = 1/x
plt.plot(x,y,color='b')
plt.grid(True)
plt.title(r'$f(x) = \frac{1}{x}$')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.show()
```

# 5. $f(x) = \frac{1}{2\sqrt{2\pi}} e^{-\frac{(x-1)^2}{8}}$
- ## x軸為漸近線
- ## 最大值為接近0.2，且與y軸交於一點

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = 1/(2*(2*np.pi)**(1/2))*np.exp(-((x-1)**2)/8)
plt.plot(x,y,color='b')
plt.grid(True)
plt.title(r'$f(x) = \frac{1}{2\sqrt{2\pi}} e^{-\frac{(x-1)^2}{8}}$')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.show()
```

# 6. $f(x) = \sqrt[3]{x^2}$
- ## y軸為對稱軸
- ## 交於原點(0,0)且為最小值0

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-15,15,100)
y = (x**2)**(1/3)
plt.plot(x,y,color='b')
plt.grid(True)
plt.title(r'$f(x) = \sqrt[3]{x^2}$')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.show()
```

# 8. $f(x) = \frac{\ln x}{x^3} $

# 9. $f(x) = 3, \;\; 1 \leq x \leq 5$

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.arange(1,6)
f = lambda x: 3*x**0
for i in x:
    plt.plot(x,f(x),color='b')
plt.grid(True)
ax = plt.gca()
ax.set_ylim([0, 3.5])
ax.set_xlim([-1, 6])
ax.spines['left'].set_position(('data',0))
ax.set_xticks([0,1,2,3,4,5,6])
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
plt.title('$f(x)=3$')
plt.show()
```

# 10. $x^2+y^2 = 1$


## 作法一

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(-1,1,100)
y = (1-x**2)**(1/2)
y1 = -(1-x**2)**(1/2)
plt.plot(x,y,x,y1,color='b')
plt.title('Parametric Equation Circle')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_aspect(1)
plt.ylim([1.5,-1.5])
plt.xlim([1.5,-1.5])
plt.show()
```

## 作法二

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 100)
r = 1
a = r*np.cos(x)
b = r*np.sin(x)
figure, axes = plt.subplots(1)
ax = plt.gca()
ax.plot(a, b)
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title('Parametric Equation Circle')
plt.ylim([1.5,-1.5])
plt.xlim([1.5,-1.5])
axes.set_aspect(1)
plt.show()
```
